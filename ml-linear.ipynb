```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sc_df = pd.read_csv('superconductors.csv')
sc_df.drop('Unnamed: 0', axis = 1, inplace = True)
sc_df
```

```python
sc_df.describe()
```

```python
correlations = sc_df.corr()
f, ax = plt.subplots(figsize= (11, 9))
cmap = sns.diverging_palette(550, 20, as_cmap = True)
sns.heatmap(correlations, cmap = cmap, linewidths = 0.5, center = 0, annot = True)
```
