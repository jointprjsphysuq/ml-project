```python
def boxplots(data):
    columns = data.columns
    size = len(columns)
    root_int = int(np.sqrt(size))
    fig = plt.figure(figsize = (15, 15))

    if root_int**2 == size:
        gs = plt.GridSpec(root_int, root_int)
    else:
        gs = plt.GridSpec(root_int + 1, root_int)
    
    ax = []

    for i in range(size):
        ax.append(fig.add_subplot(gs[i]))
    for i, col in enumerate(columns):
        r = np.random.randint(1, 10)/10
        g = np.random.randint(1, 10)/10
        b = np.random.randint(1, 10)/10
        sns.boxplot(data = data, x = col, ax = ax[i], color = (r, g, b))
        ax[i].tick_params(axis = 'both', which = 'major')
        ax[i].set_xlabel(col)
    plt.show()
```

```python
def histplots(data):
    columns = data.columns
    size = len(columns)
    root_int = int(np.sqrt(size))
    fig = plt.figure(figsize = (20, 20))

    if root_int**2 == size:
        gs = plt.GridSpec(root_int, root_int)
    else:
        gs = plt.GridSpec(root_int + 1, root_int)
    
    ax = []

    for i in range(size):
        ax.append(fig.add_subplot(gs[i]))
    for i, col in enumerate(columns):
        r = np.random.randint(1, 10)/10
        g = np.random.randint(1, 10)/10
        b = np.random.randint(1, 10)/10
        sns.histplot(data = data, x = col, ax = ax[i], color = (r, g, b), stat = 'density', bins = int(np.log(data.shape[0])/np.log(2) + 1))
        ax[i].tick_params(axis = 'both', which = 'major')
        ax[i].set_xlabel(col)
    plt.show()
```

```python
subsubset = subset_df.drop('critical_temp', axis = 1)
boxplots(subsubset)
plt.savefig('graphs/boxplots_features.pdf', dpi = 700, bbox_inches = 'tight')
```

```python
histplots(subsubset)
plt.savefig('graphs/histplots_features.pdf', dpi = 700, bbox_inches = 'tight')
```

```python
def classifier(temperature):
    for T in temperature:
        if T < 77:
            return 0
        else:
            return 1
```

```python
subset_df['Type'] = subset_df[['critical_temp']].apply(classifier, axis = 1)
subset_df['Type'].value_counts(normalize = True).plot.bar()
plt.title(r'Type frequency')
plt.ylabel('Frequency')
plt.xlabel('Type of superconductor')
plt.savefig('graphs/type-frequency.pdf', bbox_inches = 'tight')
```
